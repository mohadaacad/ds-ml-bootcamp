```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

CSV_PATH = "dataset/clean_car_dataset.csv"

df = pd.read_csv(CSV_PATH)

# Features
x = df.drop(columns=["Price", "LogPrice",'unnamed:0'],errors='ignore')

# Target
y = df["Price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)

lr_pred_log = lr.predict(x_test)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)

lr_pred_log = lr.predict(x_test)
```

```python
rf = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

rf.fit(x_train, y_train)

rf_pred_log = rf.predict(x_test)
```

```python
# Actual Price
y_test_price = np.exp(y_test)

# Linear Regression Prediction
lr_pred_price = np.exp(lr_pred_log)

# Random Forest Prediction
rf_pred_price = np.exp(rf_pred_log)
```
